```python
import pandas as pd

# Load your dataset
df = pd.read_csv('new_songs_with_details.csv')

# Drop rows where either 'file_path' or 'video_url' is missing
missing_df = df.dropna(subset=['video_url', 'file_path'])

# Optionally, save the cleaned dataset back to CSV
missing_df.to_csv('new_songs_with_details_cleaned.csv', index=False)
```

```python
import pandas as pd

# Load your dataset
new_df = pd.read_csv('new_songs_with_details_cleaned.csv', usecols=[
    'file_path', 'acousticness', 'danceability', 'energy',
    'instrumentalness', 'key', 'liveness', 'loudness',
    'speechiness', 'tempo', 'time_signature', 'valence'
])

# 1) Convert backslashes to forward slashes
new_df['file_path'] = new_df['file_path'].str.replace(r'\\', '/', regex=True)

# 2) Extract only the filename (last part of the file_path)
new_df['file_path'] = new_df['file_path'].str.split('/').str[-1]

# 3) Replace '.mp3' with '.npy' if it appears at the end
new_df['file_path'] = new_df['file_path'].str.replace(r'\.mp3$', '.npy', regex=True)

# Display the first few rows to verify
print(new_df.head())

# (Optional) Save the cleaned DataFrame
new_df.to_csv('new_songs_with_details_cleaned.csv', index=False)
```

```python
import pandas as pd

# Define the desired column order
column_order = [
    'file_path', 'valence', 'liveness', 'time_signature', 'energy',
    'speechiness', 'instrumentalness', 'acousticness', 'tempo',
    'loudness', 'danceability', 'key'
]


# Reorder the columns according to the defined order
new_df = new_df[column_order]

# Save the reordered DataFrame to a new CSV file
new_df.to_csv('Dataset_cleaned_new_proper.csv', index=False)
```
